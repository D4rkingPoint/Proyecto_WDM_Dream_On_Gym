```python
import json
import networkx as nx
import random
import matplotlib.pyplot as plt
import numpy as np

# Graficar resultados
def plot_results(demands, blocking_probabilities):
    plt.figure()
    plt.plot(demands, blocking_probabilities, marker='o')
    plt.xlabel('Número de demandas')
    plt.ylabel('Probabilidad de bloqueo')
    plt.title('Número de demandas vs Probabilidad de bloqueo')
    plt.grid(True)
    plt.show()

# Función para graficar la topología
def plot_topology(graph):
    pos = nx.spring_layout(graph)
    edge_labels = nx.get_edge_attributes(graph, 'length')

    plt.figure(figsize=(8, 6))
    nx.draw(graph, pos, with_labels=True, node_size=500, node_color="skyblue", font_weight="bold")
    nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels)
    plt.title("Topología de la red")
    plt.show()

# Clase para manejar los formatos de modulación y tasas de bits
class BitRate:
    def __init__(self, rate):
        self.rate = rate
        self.modulations = []

    def addModulation(self, modulation, fsu, max_distance):
        self.modulations.append({"modulation": modulation, "fsu": fsu, "max_distance": max_distance})

# Configuración de las tasas de bits y modulación
bitRatesDefault = []
modulation_table = [
    {"modulation": "BPSK", "fsus": [1, 4, 8, 32, 80], "distances": [4000, 4000, 4000, 4000, 4000], "bitRates": [10.0, 40.0, 100.0, 400.0, 1000.0]},
    {"modulation": "QPSK", "fsus": [1, 2, 4, 16, 40], "distances": [2000, 2000, 2000, 2000, 2000], "bitRates": [10.0, 40.0, 100.0, 400.0, 1000.0]},
    {"modulation": "8-QAM", "fsus": [1, 3, 3, 11, 27], "distances": [1000, 1000, 1000, 1000, 1000], "bitRates": [10.0, 40.0, 100.0, 400.0, 1000.0]},
    {"modulation": "16-QAM", "fsus": [1, 2, 2, 6, 20], "distances": [500, 500, 500, 500, 500], "bitRates": [10.0, 40.0, 100.0, 400.0, 1000.0]},
    {"modulation": "32-QAM", "fsus": [1, 2, 1, 6, 12], "distances": [250, 250, 250, 250, 250], "bitRates": [10.0, 40.0, 100.0, 400.0, 1000.0]},
    {"modulation": "64-QAM", "fsus": [1, 2, 1, 6, 14], "distances": [125, 125, 125, 125, 125], "bitRates": [10.0, 40.0, 100.0, 400.0, 1000.0]},
]

# Crear las tasas de bits y sus modulación
for i in range(len(modulation_table[0]["bitRates"])):
    for mod in modulation_table:
        auxB = BitRate(mod["bitRates"][i])
        auxB.addModulation(mod["modulation"], mod["fsus"][i], mod["distances"][i])
        bitRatesDefault.append(auxB)
# Función para cargar topología desde un archivo JSON
def load_topology(json_file):
    with open(json_file, 'r') as file:
        data = json.load(file)
    graph = nx.Graph()
    for link in data['links']:
        src, dst = link['src'], link['dst']
        distance = link['length']
        selected_bands = ['C', 'L', 'S', 'E'] # Cambiar esto para seleccionar las bandas a usar
        total_slots = sum(link['slots'][band] for band in selected_bands if band in link['slots'])
        graph.add_edge(src, dst, distance=distance, spectrum=[0] * total_slots)
    return graph

def select_modulation(distance, bit_rate):
    best_modulation = None
    min_fsu = float('inf')  # Inicializar con infinito para encontrar el mínimo

    for br in bitRatesDefault:
        if br.rate >= bit_rate:  # Verificar si la tasa de bits es suficiente
            for mod in br.modulations:
                if distance <= mod['max_distance']:  # Verificar la distancia
                    if mod['fsu'] < min_fsu:  # Buscar la modulación con menor FSU
                        min_fsu = mod['fsu']
                        best_modulation = mod

    if best_modulation:
        # Retornar la mejor modulación encontrada
        return best_modulation['fsu']

    return None  # Bloqueo si no hay modulación compatible

# Algoritmo Shortest Path - First Fit
def rmlsa(graph, source, destination, bit_rate, distance):
    try:
        path = nx.shortest_path(graph, source, destination, weight='distance')
    except nx.NetworkXNoPath:
        return False, []  # Bloqueo por falta de ruta

    fsu_required = select_modulation(distance, bit_rate)
    if fsu_required is None:
        return False, []  # Bloqueo por incompatibilidad de modulación

    for i in range(len(graph[path[0]][path[1]]['spectrum']) - fsu_required + 1):
        if all(graph[path[j]][path[j + 1]]['spectrum'][i:i + fsu_required] == [0] * fsu_required for j in range(len(path) - 1)):
            for j in range(len(path) - 1):
                graph[path[j]][path[j + 1]]['spectrum'][i:i + fsu_required] = [1] * fsu_required
            return True, path  # Demanda aceptada
    return False, []  # Bloqueo por falta de espectro

def weighted_random_bitrate(bitrates):
  """
  Selecciona un bitrate aleatorio de una lista de bitrates,
  dando mayor peso a los bitrates más bajos.

  Args:
    bitrates: Una lista de objetos BitRate.

  Returns:
    float: El bitrate seleccionado.
  """

  # Crear una lista de pesos, donde los bitrates más bajos tienen mayor peso
  weights = [1 / br.rate for br in bitrates]

  # Normalizar los pesos para que sumen 1
  total_weight = sum(weights)
  normalized_weights = [w / total_weight for w in weights]

  # Seleccionar un bitrate aleatorio usando los pesos
  return random.choices([br.rate for br in bitrates], weights=normalized_weights)[0]


def simulate_exponential(graph, lambda_rate, mu_rate, max_users):
    blocked = 0
    blocking_probabilities = []
    blocked_requests_info = []  # Lista para almacenar información de solicitudes bloqueadas
    blocked_modulation_summary = {}  # Diccionario para resumir bloqueos por modulación y FSU
    nodes = list(graph.nodes)

    arrival_times = np.random.exponential(1 / lambda_rate, max_users)
    current_time = 0

    for i in range(max_users):
        current_time += arrival_times[i]
        source, destination = random.sample(nodes, 2)

        # Calcular la distancia total entre los nodos seleccionados
        try:
            path = nx.shortest_path(graph, source, destination, weight='distance')
            distance = sum(graph[path[j]][path[j + 1]]['distance'] for j in range(len(path) - 1))
        except nx.NetworkXNoPath:
            distance = float('inf')  # No hay camino entre los nodos

        if distance == float('inf'):
            blocked += 1
            blocked_requests_info.append((source, destination, None, distance, None))  # No se puede calcular FSU
            continue

        bit_rate = weighted_random_bitrate(bitRatesDefault)
        fsu_required = select_modulation(distance, bit_rate)  # Obtener el FSU requerido

        success, path = rmlsa(graph, source, destination, bit_rate, distance)

        if not success:
            blocked += 1
            blocked_requests_info.append((source, destination, bit_rate, distance, fsu_required))  # Almacenar información de la solicitud bloqueada

            # Registrar la modulación y la cantidad de FSU necesarias
            if fsu_required is not None:
                mod_name = next((mod['modulation'] for br in bitRatesDefault for mod in br.modulations if mod['fsu'] == fsu_required), "Unknown")
                if (mod_name, fsu_required) in blocked_modulation_summary:
                    blocked_modulation_summary[(mod_name, fsu_required)] += 1
                else:
                    blocked_modulation_summary[(mod_name, fsu_required)] = 1

        # Mapeo de resultados cada 10 llegadas
        if (i + 1) % 10 == 0:
            blocking_prob = blocked / (i + 1)
            blocking_probabilities.append(blocking_prob)

    return blocking_probabilities, blocked_requests_info, blocked_modulation_summary


def plot_blocked_modulation_summary(blocked_modulation_summary):
    mod_names = [f"{mod} (FSU: {fsu})" for (mod, fsu) in blocked_modulation_summary.keys()]
    counts = list(blocked_modulation_summary.values())

    plt.figure(figsize=(10, 6))
    plt.bar(mod_names, counts, color='skyblue')
    plt.xlabel('Modulación y FSU requeridas')
    plt.ylabel('Número de bloqueos')
    plt.title('Resumen de bloqueos por modulación y FSU')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y')
    plt.tight_layout()  # Ajustar los márgenes para que no se corten las etiquetas
    plt.show()

# Main
if __name__ == "__main__":
    fileDirectory = '/content/drive/MyDrive/'
    topology_file = fileDirectory + "Topologias-rutas/BDM_ARPANet_4_bands.json"  # Archivo JSON con la topología
    graph = load_topology(topology_file)

    # Graficar la topología antes de la simulación
    plot_topology(graph)

    lambda_rate = 1000  # Tasa de llegada
    mu_rate = 1  # Tasa de servicio (no es utilizada en el cálculo pero puede ser relevante para análisis futuros)
    max_users = 100000  # Definir la cantidad máxima de usuarios

    blocking_probabilities, blocked_requests_info, blocked_modulation_summary = simulate_exponential(graph, lambda_rate, mu_rate, max_users)

    # Graficar resultados
    demands = list(range(10, max_users + 1, 10))
    plot_results(demands, blocking_probabilities)

    # Imprimir información de solicitudes bloqueadas
    #print("Solicitudes bloqueadas (origen, destino, bitrate, distancia, FSU requeridas):")
    #for request in blocked_requests_info:
        #print(request)  # Imprimir la información de las solicitudes bloqueadas incluyendo distancia y FSU

    # Imprimir y graficar resumen de bloqueos por modulación y FSU
    print("última probabilidad de bloqueo: ", blocking_probabilities[-1])
    print("\nResumen de bloqueos por modulación y FSU:")
    for (mod_name, fsu_count), count in blocked_modulation_summary.items():
        print(f"{mod_name} (FSU: {fsu_count}): {count} bloqueos")

    # Graficar el resumen de bloqueos
    plot_blocked_modulation_summary(blocked_modulation_summary)
```

```python
def getBand(actionBand):
    if (actionBand == 0):
        return 'C'
    elif (actionBand == 1):
        return 'L'
    elif (actionBand == 2):
        return 'S'
    elif (actionBand == 3):
        return 'E'

    return 'NoBand'
```

```python
def reward():
    global episodeRewardAccum
    global episodeIteration
    global episodeRewardsArray
    global iteration

    iteration = iteration + 1
    if (iteration % episodeIteration) == 0:
        episodeRewardsArray.append(episodeRewardAccum)
        episodeRewardAccum = 0

    value = base_env.getSimulator().lastConnectionIsAllocated()
    if (value.name == Controller.Status.Not_Allocated.name):
        #print(value.name)
        value = -1
    else:
        value = 1
    episodeRewardAccum = episodeRewardAccum + value
    return value
```

```python
def state():
    connectionEvent = base_env.getSimulator().connectionEvent
    route = base_env.getSimulator().controller.path[connectionEvent.source][connectionEvent.destination][0]
    slotsQuantity = 0
    for i in range(4):
        slotsQuantity = slotsQuantity + route[0].getSlots(getBand(i))
    obs = [0] * slotsQuantity
    slotInit = 0
    for i in range(len(route)):
        slotIterator = 0
        for j in range(4):
            slotInit = slotIterator
            for k in range(slotInit, slotInit + route[i].getSlots(getBand(j))):
                slotIterator = slotIterator + 1
                if route[i].getSlot(k-slotInit,getBand(j)):
                    obs[k] = obs[k] + 1
    return obs
```

```python
def first_fit_algorithm(src: int, dst: int, b: BitRate, c: Connection, n: Network, path, action):
    numberOfSlots = b.getNumberofSlots(0)
    link_ids = path[src][dst][0]
    c.bandSelected = "C"
    general_link = []
    for _ in range(n.getLink(0).getSlots(c.bandSelected)):
        general_link.append(False)

    for link in link_ids:
        link = n.getLink(link.id)
        link.bandSelected = c.bandSelected
        for slot in range(link.getSlots(c.bandSelected)):
            general_link[slot] = general_link[slot] or link.getSlot(slot,c.bandSelected)
    currentNumberSlots = 0
    currentSlotIndex = 0
    for j in range(len(general_link)):
        if not general_link[j]:
            currentNumberSlots += 1
        else:
            currentNumberSlots = 0
            currentSlotIndex = j + 1
        if currentNumberSlots == numberOfSlots:
            for k in link_ids:
                c.addLink(
                    k, fromSlot=currentSlotIndex, toSlot=currentSlotIndex+currentNumberSlots)
            return Controller.Status.Allocated, c
    return Controller.Status.Not_Allocated, c
```

```python
def agent_algorithm(src: int, dst: int, b: BitRate, c: Connection, n: Network, path, action):
    band = getBand(action[0])
    c.bandSelected = band
    numberOfSlots = b.getNumberofSlots(0)
    actionSpace = len(path[src][dst])
    link_ids = path[src][dst][0]
    general_link = []
    for _ in range(n.getLink(0).getSlots(band)):
        general_link.append(False)
    for link in link_ids:
        link = n.getLink(link.id)
        link.bandSelected = band
        for slot in range(link.getSlots()):
            general_link[slot] = general_link[slot] or link.getSlot(slot,band)
    currentNumberSlots = 0
    currentSlotIndex = 0
    for j in range(len(general_link)):
        if not general_link[j]:
            currentNumberSlots += 1
        else:
            currentNumberSlots = 0
            currentSlotIndex = j + 1
        if currentNumberSlots == numberOfSlots:
            for k in link_ids:
                c.addLink(
                    k, fromSlot=currentSlotIndex, toSlot=currentSlotIndex+currentNumberSlots)
            return Controller.Status.Allocated, c
    return Controller.Status.Not_Allocated, c
```

## Funciones para graficar resultados

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Funcion que muestra algunas de las stats de una funcion recompensa
def Estadisticas(rewards, pb, name):
    print("Función de recompensa ", name)
    print("Promedio = ", np.mean(rewards))
    print("Desviación = ", np.std(rewards))
    print("Minima y maxima recompensa", np.min(rewards[1:]), "-", np.max(rewards[1:]))
    print("Última probabilidad de bloqueo", pb[-1])
    return 1

# Se le entrega iteraciones, array de rewards, array de probabilidades de bloqueos, nombre funcion de recompensa, nombre agente, nombre topo
def grafico_recompensas_Bloqueo(iterations, rewards, pbs, recompensa, agent, topologia):
    data = {'X': iterations[1:], 'Y1': rewards[1:], 'Y2':pbs[1:]}
    df = pd.DataFrame(data)

    sns.set(style="whitegrid")
    f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[8, 8]))
    plt.suptitle('Funcion de Recompensa: '+ recompensa + ' - '+ agent+' - '+ topologia)

    sns.lineplot(x="X", y="Y1", data= df, ax=axs[0])
    axs[0].set_title('Recompensa v/s Iteración')
    axs[0].set_ylabel('Recompensa')
    axs[0].set_xlabel('Iteración')

    sns.lineplot(x="X", y="Y2", data= df, ax=axs[1])
    axs[1].set_title('Prob. Bloqueo v/s Iteración')
    axs[1].set_ylabel('Prob. Bloqueo')
    axs[1].set_xlabel('Iteración')

    f.tight_layout()
    Estadisticas(rewards, pbs, recompensa)
    plt.show()

def plot_block_FSU(blocked_bit_rates, bit_rates_dict):
    bit_rate_names = [
        f"{bit_rates_dict[bit_rate_index]['modulation']} (FSU: {bit_rates_dict[bit_rate_index]['FSU']})"
        for bit_rate_index in blocked_bit_rates.keys()
    ]
    counts = list(blocked_bit_rates.values())

    plt.figure(figsize=(10, 6))
    plt.bar(bit_rate_names, counts, color='skyblue')
    plt.xlabel('Modulación (FSU)')
    plt.ylabel('Número de bloqueos')
    plt.title('Resumen de bloqueos por Modulación y FSU')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

def graficar_probabilidad_bloqueo(prob_bloqueo, iterationAccum):
  """
  Grafica la probabilidad de bloqueo del sistema a lo largo del tiempo.

  Args:
    prob_bloqueo: Una lista que contiene las probabilidades de bloqueo
                   en diferentes puntos de tiempo.
    iterationAccum: El número de iteraciones que se acumulan antes de
                    calcular cada punto de la probabilidad de bloqueo.
  """

  # Crea una lista de puntos en el eje x (tiempo)
  x = [i * iterationAccum for i in range(len(prob_bloqueo))]

  # Crea el gráfico
  plt.plot(x, prob_bloqueo)

  # Añade etiquetas y título
  plt.xlabel('Iteraciones')
  plt.ylabel('Probabilidad de Bloqueo')
  plt.title('Probabilidad de Bloqueo del Sistema')

  # Muestra el gráfico
  plt.show()
```

```python
def rewardLengthBased():
    global episodeRewardAccum
    global episodeIteration
    global episodeRewardsArray
    global iteration

    iteration = iteration + 1
    if (iteration % episodeIteration) == 0:
        episodeRewardsArray.append(episodeRewardAccum)
        episodeRewardAccum = 0

    connectionsNow = base_env.getSimulator().controller.connections
    controllerNow = base_env.getSimulator().controller

    value = base_env.getSimulator().lastConnectionIsAllocated()

    if (value.name == Controller.Status.Not_Allocated.name):
        value = -1
    else:

        src = connectionsNow[-1].links[0].src
        dst = connectionsNow[-1].links[-1].dst
        actualPath = connectionsNow[-1].links
        actualPathLength = 0
        for link in actualPath:
          actualPathLength += link.length
        # Possible paths
        paths = controllerNow.path[src][dst]
        PathsLength = []
        for path in paths:
          linksCumLength = 0
          for link in path:
            linksCumLength += link.length
          PathsLength.append(linksCumLength)

        if (actualPathLength <= PathsLength[0]):
          value = 1
        elif (actualPathLength <= PathsLength[1] or actualPathLength <= PathsLength[2]):
          value = 0.9
        else:
          value = 0.8

    episodeRewardAccum = episodeRewardAccum + value
    return value
```

###Función de recompensa basado en la selección de banda


```python
def rewardBandSelect():
    global episodeRewardAccum
    global episodeIteration
    global episodeRewardsArray
    global iteration

    iteration = iteration + 1
    if (iteration % episodeIteration) == 0:
        episodeRewardsArray.append(episodeRewardAccum)
        episodeRewardAccum = 0

    reward = 0
    connectionsNow = base_env.getSimulator().controller.connections

    value = base_env.getSimulator().lastConnectionIsAllocated()
    if (value.name == Controller.Status.Not_Allocated.name):
        reward -= 1.0
    elif connectionsNow[-1].bandSelected == "C" or connectionsNow[-1].bandSelected == "L" or connectionsNow[-1].bandSelected == "S":
        reward +=1.0
    elif connectionsNow[-1].bandSelected == "E":
        reward +=0.8

    episodeRewardAccum = episodeRewardAccum + reward
    return reward
```
